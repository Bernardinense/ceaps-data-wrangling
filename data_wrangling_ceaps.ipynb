```python
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_columns', None) # Mostrar todas as colunas ao exibir DataFrames
pd.set_option('display.max_colwidth', 50) # N√£o truncar o conte√∫do das colunas ao exibir DataFrames
pd.set_option('display.float_format', '{:.2f}'.format) #Formato dos numeros


#Carregar o dataset CEAPS - carrega somente um Dataset

# df = pd.read_csv('despesa_ceaps_2022.csv',
#     sep=';',
#     encoding='latin-1',
#     skiprows=1
# )
# print("Dataset carregado com sucesso")

#Para carregar quantos datasets quiser = forma automatica


anos = [2019, 2020, 2021, 2022]
lista_dfs = []

for ano in anos:
    arquivo = f'despesa_ceaps_{ano}.csv'
    df_ano = pd.read_csv(arquivo, sep=';', encoding='latin-1', skiprows=1)
    lista_dfs.append(df_ano)
    print(f"‚úÖ {ano}: {len(df_ano):,} registros")

df = pd.concat(lista_dfs, ignore_index=True)
print(f"\nüìä Total: {len(df):,} registros")

```

```python
#Limpeza e Transforma√ß√£o

#Criar copia do dataframe para n√£o alterar coisas = BOA PRATICA
```

```python
df_clean = df.copy()
print("‚úÖ C√≥pia criada: df_clean")
```

```python
# Converter valor para num√©rico coluna VALOR REEMBOLSADO

df_clean['VALOR_REEMBOLSADO'] = (
    df_clean['VALOR_REEMBOLSADO']
    .str.replace(',', '.')  # Trocar v√≠rgula por ponto
    .astype(float)          # Converter para float
)

print(f"Tipo ap√≥s convers√£o: {df_clean['VALOR_REEMBOLSADO'].dtype}")
print(f"\nEstat√≠sticas do VALOR_REEMBOLSADO:")
print(df_clean['VALOR_REEMBOLSADO'].describe())
```

```python
# Converter para datetime coluna DATA
df_clean['DATA'] = pd.to_datetime(
    df_clean['DATA'], 
    format='%d/%m/%Y',
    errors='coerce'  # Valores inv√°lidos viram NaT (Not a Time)
)

print(f"Tipo ap√≥s convers√£o: {df_clean['DATA'].dtype}")
print(f"\nData mais antiga: {df_clean['DATA'].min()}")
print(f"Data mais recente: {df_clean['DATA'].max()}")
```

```python
# Verificar se houve erros na convers√£o (NaT)
datas_invalidas = df_clean['DATA'].isna().sum()
print(f"Datas que n√£o puderam ser convertidas: {datas_invalidas}")
```

```python
# Extrair componentes da data
df_clean['DIA'] = df_clean['DATA'].dt.day
df_clean['MES_NOME'] = df_clean['DATA'].dt.month_name()
df_clean['DIA_SEMANA'] = df_clean['DATA'].dt.day_name()
df_clean['TRIMESTRE'] = df_clean['DATA'].dt.quarter

print("‚úÖ Colunas auxiliares criadas: DIA, MES_NOME, DIA_SEMANA, TRIMESTRE")
df_clean[['DATA', 'DIA', 'MES_NOME', 'DIA_SEMANA', 'TRIMESTRE']].head()
```

```python
# Fun√ß√£o para limpar CNPJ/CPF (manter apenas n√∫meros)
def limpar_documento(doc):
    if pd.isna(doc):
        return None
    return re.sub(r'\D', '', str(doc))  # Remove tudo que n√£o √© d√≠gito

# Fun√ß√£o para formatar CNPJ/CPF
def formatar_documento(doc):
    if pd.isna(doc) or doc == '':
        return None
    
    doc_limpo = re.sub(r'\D', '', str(doc))
    
    if len(doc_limpo) == 11:  # CPF
        return f"{doc_limpo[:3]}.{doc_limpo[3:6]}.{doc_limpo[6:9]}-{doc_limpo[9:]}"
    elif len(doc_limpo) == 14:  # CNPJ
        return f"{doc_limpo[:2]}.{doc_limpo[2:5]}.{doc_limpo[5:8]}/{doc_limpo[8:12]}-{doc_limpo[12:]}"
    else:
        return doc_limpo  # Retorna como est√° se n√£o for CPF nem CNPJ v√°lido

# Aplicar fun√ß√µes
df_clean['CNPJ_CPF_LIMPO'] = df_clean['CNPJ_CPF'].apply(limpar_documento)
df_clean['CNPJ_CPF_FORMATADO'] = df_clean['CNPJ_CPF'].apply(formatar_documento)

print("‚úÖ Colunas criadas: CNPJ_CPF_LIMPO e CNPJ_CPF_FORMATADO")
df_clean[['CNPJ_CPF', 'CNPJ_CPF_LIMPO', 'CNPJ_CPF_FORMATADO']].head(10)
```

```python
# Criar coluna identificando se √© CPF ou CNPJ
def tipo_documento(doc_limpo):
    if pd.isna(doc_limpo):
        return None
    if len(str(doc_limpo)) == 11:
        return 'CPF'
    elif len(str(doc_limpo)) == 14:
        return 'CNPJ'
    else:
        return 'INVALIDO'

df_clean['TIPO_DOCUMENTO'] = df_clean['CNPJ_CPF_LIMPO'].apply(tipo_documento)

print("Distribui√ß√£o por tipo de documento:")
print(df_clean['TIPO_DOCUMENTO'].value_counts())
```

```python
# Tratar valores nulos na Coluna DETALHAMENTO
```

```python
# Verificar nulos em DETALHAMENTO
nulos_detalhamento = df_clean['DETALHAMENTO'].isna().sum()
vazios_detalhamento = (df_clean['DETALHAMENTO'] == '').sum()

print(f"Valores nulos em DETALHAMENTO: {nulos_detalhamento}")
print(f"Valores vazios em DETALHAMENTO: {vazios_detalhamento}")
```

```python
# Substituir vazios e nulos por 'N√£o informado'
df_clean['DETALHAMENTO'] = df_clean['DETALHAMENTO'].replace('', np.nan)
df_clean['DETALHAMENTO'] = df_clean['DETALHAMENTO'].fillna('N√£o informado')

print("‚úÖ Valores vazios/nulos substitu√≠dos por 'N√£o informado'")
```

```python
#Limpar e padronizar nome dos fornecedores
```

```python
# Remover espa√ßos extras e padronizar mai√∫sculas
df_clean['FORNECEDOR'] = (
    df_clean['FORNECEDOR']
    .str.strip()           # Remove espa√ßos in√≠cio/fim
    .str.upper()           # Tudo mai√∫sculo
    .str.replace(r'\s+', ' ', regex=True)  # M√∫ltiplos espa√ßos ‚Üí um espa√ßo
)

print("‚úÖ Nomes de fornecedores padronizados")
```

```python
#Padronizar nomes dos senadores
```

```python
# Verificar se h√° inconsist√™ncias nos nomes
df_clean['SENADOR'] = (
    df_clean['SENADOR']
    .str.strip()
    .str.upper()
    .str.replace(r'\s+', ' ', regex=True)
)

print(f"Total de senadores ap√≥s padroniza√ß√£o: {df_clean['SENADOR'].nunique()}")
```

```python
#Remover duplicados
```

```python
duplicados_antes = df_clean.duplicated().sum()
print(f"Duplicados antes: {duplicados_antes}")

if duplicados_antes > 0:
    df_clean = df_clean.drop_duplicates()
    print(f"‚úÖ {duplicados_antes} duplicados removidos")
else:
    print("‚úÖ Nenhum duplicado encontrado")
```
